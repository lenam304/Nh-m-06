```
%cd /content/gdrive/MyDrive/darknet
 
import glob2
import os
 
files_test = []
files_train = []
for ext in [".png", ".jpeg", "*.jpg"]:
  image_files = glob2.glob(os.path.join("/content/gdrive/MyDrive/darknet/data/UIT-VD1_Copy/Test", ext))
  files_test += image_files
for ext in [".png", ".jpeg", "*.jpg"]:
  image_files = glob2.glob(os.path.join("/content/gdrive/MyDrive/darknet/data/UIT-VD1_Copy/Train", ext))
  files_train += image_files

# Tạo file test.txt
with open("test.txt", "w") as f:
  for idx in files_test:
      f.write(idx+'\n')
 
# Tạo file train.txt
with open("train.txt", "w") as f:
  for idx in files_train:
      f.write(idx+'\n')
```

```
%cd /content/gdrive/MyDrive/darknet
!mkdir backup
!echo classes=8 > yolo.data
!echo train=train.txt >> yolo.data
!echo valid=test.txt >> yolo.data
!echo names=yolo.names >> yolo.data
!echo backup=backup >> yolo.data
```
